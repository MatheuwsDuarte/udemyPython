```python
import pandas as pd
```

```python
df = pd.read_csv('house.csv')
```

```python
df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Matheus'
df
```

```python
df.loc[df['SellerG'] == 'Matheus', ['Method', 'Type']] = 'Pending'
df
```

```python
df.groupby(['SellerG']).mean(numeric_only=True) #media dos vendedores
```

```python
df.groupby(['SellerG']).sum(numeric_only=True).sort_values('Price', ascending= False).head(10) #Soma as vendas dos vendedores e mostra o top10
```
